# Scraping LinkedIn

```python
from robobrowser import RoboBrowser
from bs4 import BeautifulSoup
import requests
```

```python
br = RoboBrowser()
```

```python
br.open("https://connectome-qasimwani.c9users.io")
```

```python
form = br.get_forms()
form = form[0]
```

```python
form['username'] = 'elonmusk'
form['password'] = 'qasimwani'
```

```python
br.submit_form(form)
```

```python
print(br.url)
```

```python
content = str(br.parsed())
print(content)
```

```python
contentPost = 'I have hacked the site. All of these posts were written by a script file in python using RoboBrowser Web Scraping Package. **Insert Evil Laugh Here...**'
imageID = 'https://videohive.img.customer.envatousercontent.com/files/224212193/You%20Have%20Been%20Hacked%20590x332.jpg?auto=compress%2Cformat&fit=crop&crop=top&max-h=8000&max-w=590&s=9dcc639711b6e066b0a18abe577e8f5b'
for x in range(0,10):
    br.open(br.url, method='post', data={'title':'What can I do???','description':contentPost,'image':imageID})
```

```python
newContent = str(br.parsed(features='lxml'))
print(newContent)
```

```python
soup = BeautifulSoup(newContent,'html.parser')
```

```python
results = soup.find_all('div',attrs={'class':'posts rounded'})
```
